<h2><b>IV. Data Cleaning</b></h2>

```python
# Filter dataset asli dan dimasukkan ke variabel df_filter
df_filter = df[df['Quantity'] > 0]
```

```python
# Membuat grafik analisis korelasi dari kolom Quantity dan Price
sns.lmplot(data=df_sampel,x='Quantity', y='Price')
```

Dari hasil visualisasi analisis korelasi kolom Price dan Quantity maka dapat diambil kesimpulan bahwa garis regresi menunjukkan sedikit kemiringan ke bawah yang mengindikasikan bahwa harga produk cenderung lebih rendah ketika jumlah pembelian produk lebih besar (meningkat), namun banyak titik terkonsentrasi di kiri bawah sehingga menunjukkan bahwa sebagian besar transaksi memiliki harga dan jumlah pembelian yang kecil dan terdapat outlier di bagian kanan (jumlah pembelian produk yang besar) dan di bagian atas (harga produk yang sangat tinggi dibanding produk lain)

```python
# Menggunakan uji korelasi spearman correlation karena memiliki data skewed/tidak normal pada kolom Price dan Quantity
cor_rho, pval_s = stats.spearmanr(df_sampel['Quantity'], df_sampel['Price'])

print(f"rho-correlation: {cor_rho:.2f}, p-value: {pval_s}")
```

Hasil uji korelasi menggunakan spearmanr menunjukkan bahwa terdapat korelasi negatif lemah hingga sedang yaitu -0.31 artinya semakin tinggi harga produk maka jumlah pembelian cenderung lebih sedikit, sedangkan untuk p-value yaitu 2.1474484492865098e-108 (0. 107nol 2147) lebih kecil dari 0.05 maka menunjukkan bahwa terdapat yang hubungan signifikan antara harga produk dan jumlah pembelian.

6. Apakah terdapat perbedaan harga produk yang signifikan berdasarkan negara pembeli?

```python
# Mengelompokkan negara dengan harga produknya menggunakan for loop dan menyaring negara menggunakan .unique() yang kemudian dimasukkan ke variabel grup_negara_hargaproduk
grup_negara_hargaproduk = []
for negara in df_sampel['Country'].unique():
    grup_negara_hargaproduk.append(df_sampel[df_sampel['Country'] == negara]['Price'])
```

H0: Tidak terdapat perbedaan harga produk yang signifikan berdasarkan negara pembeli

H1: Terdapat perbedaan harga produk yang signifikan berdasarkan negara pembeli

```python
# Uji ANOVA
res = stats.f_oneway(*grup_negara_hargaproduk)
print('P-value:',res.pvalue)
```

 Dari hasil uji ANOVA diatas dapat diketahui bahwa  p-value lebih dari 0.05 yang artinya gagal menolak atau dengan kata lain menerima H0, sehingga tidak terdapat perbedaan harga produk yang signifikan berdasarkan negara pembeli. Oleh karena itu, dapat dikatakan bahwa value harga produk di negara-negara yang terdapat pelanggan yang membeli produk perusahaan tidak memiliki perbedaan harga signifikan.
